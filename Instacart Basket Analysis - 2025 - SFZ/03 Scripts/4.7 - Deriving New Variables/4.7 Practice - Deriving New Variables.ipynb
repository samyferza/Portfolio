#### If-Statements with User-Defined Functions

```python
ords_prods_merge.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
# create smaller subset of data
df = ords_prods_merge[:1000000]
```

```python
# function for categorizing prices
def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# create new column with price_label output
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# return value counts of price_range
df['price_range'].value_counts(dropna = False)
```

```python
# return max value of prices
df['prices'].max()
```

#### If-Statements with the loc() Function

```python
# create high range label
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# create mid range label
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# create low range label
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# return value counts of price_range_loc
df['price_range_loc'].value_counts(dropna = False)
```

```python
# create high range label on full ords_prods_merge
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# create mid range label on full ords_prods_merge
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# create low range label on full ords_prods_merge
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# return value counts of price_range_loc
ords_prods_merge['price_range_loc'].value_counts(dropna = False)
```

#### If-Statements with For-Loops

```python
# example
for x in range(30,45):
    print("My age is %d" % (x))
```

```python
# frequency of orders_day_of_week column
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# assign busy-ness label based on orders_day_of_week
result = []
for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# check output
result
```

```python
# add result to ords_prods_merge as a new column
ords_prods_merge['busiest_day'] = result
```

```python
# check value count of new column
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

```python
# create label for regular days
ords_prods_merge.loc[ords_prods_merge['orders_day_of_week'].isin([6, 2, 5]), 'busiest_days'] = 'Regularly busy Days'
```

```python
# create label for slowest days
ords_prods_merge.loc[ords_prods_merge['orders_day_of_week'].isin([3, 4]), 'busiest_days'] = 'Slowest Days'
```

```python
# create list of hours sorted by value count
hours_sorted = ords_prods_merge['order_hour_of_day'].value_counts(dropna = False).index.to_numpy()
```

```python
# create empty list and for-loop to populate list with busiest time of day labels
busiest_hour = []
for row in ords_prods_merge['order_hour_of_day']:
    if row in hours_sorted[:8]:
        busiest_hour.append("Most orders")
    elif row in hours_sorted[8:16]:
        busiest_hour.append("Average orders")
    else:
        busiest_hour.append("Fewest orders")
```

```python
# create column in dataframe equal to busiest_hour
ords_prods_merge['busiest_period_of_day'] = busiest_hour
```
